# 1. 이미지 데이터에서 픽셀 데이터 가져오기

```python
def get_pixel(fname, size = 64): 
    img = Image.open(fname) #이미지 데이터 열기
    img = img.convert('L') #그레이스케일로 변환하기
    img = img.resize((size,size), Image.ANTIALIAS) #리사이즈하기. 28*28로
    pixel_data = img.getdata() #픽셀데이터 가져오기
    pixels = np.array(pixel_data) #numpy 배열로 변환하기
    return pixels
```

# 2. train_set 추출 

```python
# label=0에 해당하는 DataFrame 형성
label_num = 0
df= pd.DataFrame()
for i in range(0,103):
    a = [str(i),'.png'] # 파일 이름
    a = ''.join(a)
    arr1 = get_pixel(a) 
    df1 = pd.DataFrame(arr1).T
    df1['label'] = label_num 
    df = df.append(df1,ignore_index=True)
    print(df)

# label=1에 해당하는 DataFrame 형성
label_num = 1
for i in range(200,305):
    a = [str(i),'.png']
    a = ''.join(a)
    arr1 = get_pixel(a)
    df1 = pd.DataFrame(arr1).T
    df1['label'] = label_num
    df = df.append(df1,ignore_index=True)
    print(df)
   
# label=2에 해당하는 DataFrame 형성
label_num = 2
for i in range(400,483):
    a = [str(i),'.png']
    a = ''.join(a)
    arr1 = get_pixel(a)
    df1 = pd.DataFrame(arr1).T
    df1['label'] = label_num
    df = df.append(df1,ignore_index=True)
    print(df)    

# label=3에 해당하는 DataFrame 형성
label_num = 3
for i in range(600,705):
    a = [str(i),'.png']
    a = ''.join(a)
    arr1 = get_pixel(a)
    df1 = pd.DataFrame(arr1).T
    df1['label'] = label_num
    df = df.append(df1,ignore_index=True)
    print(df)
    
df.to_csv('train.csv', index=False) #형성된 DataFrame을 CSV 파일로
```
