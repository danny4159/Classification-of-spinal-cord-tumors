# 1. get_pixel 함수 정의

```python
# 이미지 데이터에서 픽셀 데이터 가져오기
def get_pixel(fname, size = 64): # 28, 32, 64, 128, 224(googlenet)
    img = Image.open(fname) # 이미지 데이터 열기
    img = img.convert('L') # RGB: 삼원색. L: 그레이스케일
    img = img.resize((size,size), Image.ANTIALIAS) # 리사이즈하기. 28*28로
    pixel_data = img.getdata() # 픽셀데이터 가져오기
    pixels = np.array(pixel_data) # numpy 배열로 변환하기
    return pixels
```

# 2. train_dataset 추출 

```python
# label=0에 해당하는 DataFrame 형성
label_num = 0
df= pd.DataFrame()
for i in range(1,201):
    a = [str(i),'.png'] # 파일 이름
    a = ''.join(a)
    arr1 = get_pixel(a) 
    df1 = pd.DataFrame(arr1).T
    df1['label'] = label_num 
    df = df.append(df1,ignore_index=True)
    print(df)

```
