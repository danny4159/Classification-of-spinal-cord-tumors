# 0. 모듈 임포트

```python
pip install tensorflow keras
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import numpy as np

# 랜덤시드 고정시키기
np.random.seed(3)
```

# 1. ImageDataGenerator 객체 생성

```python
datagen = ImageDataGenerator(
        rotation_range=25,      # 무작위 회전의 각도 범위
        shear_range=0.1,        # 층밀리기의 강도
        zoom_range=0.1)         # 무작위 줌의 범위
```

# 2. Augmentation 실행

```python
# label=0
for i in range(1,21):
    a = [str(i),'.png']
    a = ''.join(a)
    img = load_img(a) 
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)
    n = 0
    for batch in datagen.flow(x, batch_size=1,save_to_dir='preview', save_prefix=str(n), save_format='png'):
        n += 1
        if n > 10:
            break
```
